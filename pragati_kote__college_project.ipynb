```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
data=pd.read_csv("/content/student college data (1) (2).csv")
data.head(10)
data['type_school']=data['type_school'].apply({'Academic':1,'Vocational':0}.get)
data['school_accreditation']=data['school_accreditation'].apply({'A':1,'B':2}.get)
data['gender']=data['gender'].apply({'Male':1,'Female':0}.get)
data['interest']=data['interest'].apply({'Less Interested':0,'Very Interested':1,'Uncertain':2,'Not Interested':3,'Quiet Interested':4}.get)
data['residence']=data['residence'].apply({'Urban':1,'Rural':0}.get)
data['parent_was_in_college']=data['parent_was_in_college'].astype('int64')
```
